Titanic data set problem (From Keggle) -

We start by importing all the necessary libraries.

```python
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd;
from sklearn.ensemble import RandomForestClassifier as randFc;
from sklearn.metrics import accuracy_score;
```

Now, we import the data from the csv files using pandas.

```python
# The train data set we are given
train = pd.read_csv('train.csv');
# The data set we gotta test
test = pd.read_csv('test.csv');
# show data
train.head()
```

```python
col_surv=train.pop('Survived');
col_surv.head()
```

```python
col_numeric = list(train.dtypes[train.dtypes != "object"].index);
train[col_numeric].head()
```

```python
train['Age'].fillna(train.Age.mean(), inplace=True);
train.describe()
```

```python
train.head()
```

```python
model = randFc(n_estimators=100)
model.fit(train[col_numeric], col_surv)
```

```python
print("Train Accuracy = " + str(accuracy_score(col_surv, model.predict(train[col_numeric]))));
```

```python
test['Survived'] = 4.0;
test[col_numeric].head()
```

```python
test['Age'].fillna(test.Age.mean(), inplace=True)
```

```python
test = test[col_numeric].fillna(test.mean()).copy()
test.describe()
```

```python
predi = model.predict(test)
predi
```

```python
submission = pd.DataFrame({
    "PassengerId": test['PassengerId'],
    "Survived": predi
})
submission.to_csv('titanic_pred.csv', index=False)
```
